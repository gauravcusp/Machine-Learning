### Question 1 (40 points)

In this question, you will model traffic counts in Pittsburgh using Gaussian process (GP) regression.  The included dataset, "PittsburghTrafficCounts.csv", represents the average daily traffic counts computed by traffic sensors at over 1,100 locations in Allegheny County, PA.  The data was collected from years 2012-2014 and compiled by Carnegie Mellon Universityâ€™s Traffic21 Institute; we have the longitude, latitude, and average daily count for each sensor.  

Given this dataset, your goal is to learn a model of traffic count as a function of spatial location.  To do so, fit a Gaussian Process regression model to the observed data.  While you can decide on the precise kernel specification, you should try to achieve a good model fit, as quantified by a log marginal likelihood value greater than (i.e., less negative than) -1400.  Here are some hints for getting a good model fit:

* We recommend that you take the logarithm of the traffic counts, and then subtract the mean of this vector, before fitting the model.

* Since the data is noisy, don't forget to include a noise term (WhiteKernel) in your model.

* When fitting a GP with RBF kernel on multidimensional data, you can learn a separate length scale for each dimension, e.g., length_scale=(length_scale_x, length_scale_y). 

Your Python code should provide the following five outputs:

1) The kernel after parameter optimization and fitting to the observed data. (10 pts)

2) The log marginal likelihood of the training data. (5 pts)

3) Show a 2-D plot of the model's predictions over a mesh grid of longitude/latitude (with color corresponding to the model's predictions) and overlay a 2-D scatter plot of sensor locations (with color corresponding to the observed values). (10 pts)

4) What percentage of sensors have average traffic counts more than two standard deviations higher or lower than the model predicts given their spatial location? (5 pts)

5) Show a 2-D scatter plot of the sensor locations, with three colors corresponding to observed values a) more than two standard deviations higher than predicted, b) more then two standard deviations lower than predicted, and c) within two standard deviations of the predicted values. (10 pts)


```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel,WhiteKernel
from sklearn.model_selection import train_test_split
data1=pd.read_csv("PittsburghTrafficCounts.csv")
data1.head()
```

```python
data1['log_trafficCount'] = np.log(data1.AvgDailyTrafficCount)
```

```python
data1['log_trafficCount'] = data1['log_trafficCount'] - np.mean(data1.log_trafficCount)
```

```python
data1.head()
```

```python
kernel = 1.0 * RBF(length_scale = 1.0, length_scale_bounds=(1e-3, 1e3))+\
                WhiteKernel(noise_level=1., noise_level_bounds=(1e-10, 1e10))

gpc_rbf = GaussianProcessRegressor(kernel=kernel).fit(data1.iloc[:,0:2], data1.log_trafficCount)

x_min, x_max = min(data1.Longitude), max(data1.Longitude)
y_min, y_max = min(data1.Latitude), max(data1.Latitude)

xx, yy = np.meshgrid(np.linspace(x_min, x_max),
                     np.linspace(y_min, y_max))
```

#### Part 1: The kernel after parameter optimization and fitting to the observed data.

```python
print('The kernel after parameter optimization and fitting to the observed data: \n{}'.format(gpc_rbf.kernel_))
```

#### Part 2: The log marginal likelihood of the training data

```python
print('The log marginal likelihood of the training data: \n{}'.format(gpc_rbf.log_marginal_likelihood_value_))
```

#### Part 3: Show a 2-D plot of the model's predictions over a mesh grid of longitude/latitude (with color corresponding to the model's predictions) and overlay a 2-D scatter plot of sensor locations (with color corresponding to the observed values).

```python
plt.figure(figsize=(5, 5))
    # Plot the predicted probabilities. For that, we will assign a color to
    # each point in the mesh [x_min, m_max]x[y_min, y_max].
plt.subplot(1, 1, 1)
Z = gpc_rbf.predict(np.c_[xx.ravel(), yy.ravel()])
# Put the result into a color plot
Z = Z.reshape((xx.shape[0], xx.shape[1]))
plt.imshow(Z, extent=(x_min, x_max, y_min, y_max), origin="lower")

# Plot also the training points
plt.scatter(data1.iloc[:,0],data1.iloc[:,1], c=np.array(["r", "g", "b"])[data1.log_trafficCount])
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks(())
plt.yticks(())
plt.title("%s, LML: %.3f" %
          ("RBF", gpc_rbf.log_marginal_likelihood(gpc_rbf.kernel_.theta)))

plt.tight_layout()
plt.show()
```

##### The plot shows the locations of sensors depending on their location. The blue color here represents the potential outliers in the data.

#### Part 4: What percentage of sensors have average traffic counts more than two standard deviations higher or lower than the model predicts given their spatial location?

```python
X = data1.iloc[:,0:2]
y = data1.log_trafficCount
```

```python
y_pred, sigma = gpc_rbf.predict(X, return_std=True)
```

```python
div = (y_pred - y) / (2. * sigma + 1e-8)
```

```python
print('Percent of sensors having counts more/less than 2 standard deviations is: {:.1%}'.format((np.abs(div)>1).mean()))
```

#### Part 5: Show a 2-D scatter plot of the sensor locations, with three colors corresponding to observed values a) more than two standard deviations higher than predicted, b) more then two standard deviations lower than predicted, and c) within two standard deviations of the predicted values

```python
data1['div'] = (y_pred - y) / (2. * sigma + 1e-8)
```

```python
plt.figure(figsize=(5, 5))
    # Plot the predicted probabilities. For that, we will assign a color to
    # each point in the mesh [x_min, m_max]x[y_min, y_max].
plt.subplot(1, 1, 1)
Z = gpc_rbf.predict(np.c_[xx.ravel(), yy.ravel()])
# Put the result into a color plot
Z = Z.reshape((xx.shape[0], xx.shape[1]))
plt.imshow(Z, extent=(x_min, x_max, y_min, y_max), origin="lower")

# Plot also the training points
plt.scatter(data1.iloc[:,0],data1.iloc[:,1], c=np.array(["r", "g", "b"])[data1['div']])
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks(())
plt.yticks(())
plt.title("%s, LML: %.3f" %
          ("RBF", gpc_rbf.log_marginal_likelihood(gpc_rbf.kernel_.theta)))

plt.tight_layout()
plt.show()
```

##### The plot here represents three categories of data. The blue points indicate the sensors tracking avg vehicle count more than 2 standard deviations, the green points represent ave vehicle counts less than 2 standard deviations while the red points indicate the avg vehicle count in between the two thresholds.

### Part 1:  Learn a Bayesian network structure using only the six features ["'\% Smokers'","'\% Obese'","'Violent Crime Rate'","'80/20 Income Ratio'","'\% Children in Poverty'","'Average Daily PM2.5'"]. Use pd.cut() to discretize each feature into 5 categories: 0,1,2,3,4.

```python
data2 = data2[["County","'\% Smokers'","'\% Obese'","'Violent Crime Rate'","'80/20 Income Ratio'","'\% Children in Poverty'","'Average Daily PM2.5'"]]
data2.head()
```

```python
## You will get an error but keep scorlling

for i in data2.columns[1:]:
    pd.to_numeric(data2[i])
```

```python
data2 = data2.replace('?',0)
```

```python
#for i in data2.columns[1:]:
#    pd.to_numeric(data2[i])
data2 = data2.convert_objects(convert_numeric=True)
data2.dropna(axis=0, inplace=True)
```

```python
## Using qcut instead of cut because I was getting error in pgmpy package with cut method
for i in data2.columns[1:]:
    data2[i] = pd.qcut(data2[i],q=5,labels=False)
data2.head()
```

#### a) Use HillClimbSearch and BicScore to learn the Bayesian network structure

```python
from pgmpy.estimators import HillClimbSearch
from pgmpy.estimators import BicScore
data = data2.iloc[:,1:]
hc = HillClimbSearch(data, scoring_method=BicScore(data))
best_model = hc.estimate()
print(best_model.edges())
```

```python
from pgmpy.models import BayesianModel
from pgmpy.estimators import MaximumLikelihoodEstimator

model = BayesianModel(best_model.edges())
model.fit(data, estimator=MaximumLikelihoodEstimator)
for cpd in model.get_cpds():
    print("CPD of {variable}:".format(variable=cpd.variable))
    print(cpd)
```

#### b) Which 5 counties have the lowest (most negative) log-likelihood values? Please show a ranked list of the top counties' names and log-likelihood values.

```python
## Using function used in lab
# Log-likelihood of data record x given model
def LL(x,model,verbose=False):
    loglike = 0
    for cpd in model.get_cpds():
        temp_cpd = cpd.copy()
        thevariable = temp_cpd.variable
        theparents = model.predecessors(thevariable)
        for parent in theparents:
            temp_cpd.reduce([(parent, x[parent])])
        theprob = temp_cpd.get_values()[x[thevariable],0]
        if verbose:
            print (thevariable,theparents,theprob)
        loglike += np.log(theprob)
    return loglike
```

```python
exmp=data2.apply(lambda x: LL(x,model),axis=1)
```

```python
import matplotlib.pylab as plt
plt.figure(figsize=(20,4))
plt.scatter(range(len(exmp)),exmp)
plt.xlim(-10,520)
plt.plot(exmp,"r--")
plt.show()
```

```python
print (exmp.sort_values().head(5))
```

```python
pd.concat([data2['County'], exmp], axis=1).sort_values([0], ascending=True).head(5)
```

### Part 2: Cluster based anomaly detection. Use all numeric features for this part, and do not discretize.

#### (a) Clustering with k-means. Please use k=3 clusters. Compute each record's distance to the nearest cluster center and report the five counties which have the longest distances. (10 pts)

```python
from sklearn.cluster import KMeans
data2=pd.read_csv("2016CountyHealthIndicators.csv")
data2 = data2.replace('?',0)
data2 = data2.convert_objects(convert_numeric=True)
data2.dropna(axis=0, inplace=True)
data2.iloc[:,1:] = (data2.iloc[:,1:] - data2.iloc[:,1:].mean()) / data2.iloc[:,1:].std()
```

```python
# Use k-means package from sklearn
KM=KMeans(n_clusters=3,random_state=111)
scor=KM.fit_predict(data2.iloc[:,1:])

# distance to each cluster center
res=pd.DataFrame(KM.transform(data2.iloc[:,1:]))
res.head()
```

```python
from sklearn.cluster import KMeans
KM=KMeans(n_clusters=5,random_state=111)
KM.fit(data2.iloc[:,1:])
res=pd.DataFrame(KM.transform(data2.iloc[:,1:]))
res=pd.concat((pd.DataFrame(data2.iloc[:,0]),res.min(1)),axis=1)
res.columns=["County","score"]
print (res.sort_values("score",ascending=False)[:5])
print

# also check for tiny clusters
print (pd.Series(KM.predict(data2.iloc[:,1:])).value_counts())
```

#### (b) Cluster with Gaussian Mixture. Please repeat (2)a but use log-likelihood for each record (rather than distance) as the measure of anomalousness.

```python
from sklearn.mixture import GaussianMixture
GM=GaussianMixture(n_components=3,random_state=111)
GM.fit(data2.iloc[:,1:])
```

```python
len(GM.score_samples(data2.iloc[:,1:]))
```

```python
import matplotlib.pylab as plt
X=data2.iloc[:,1:]
plt.figure(figsize=(20,4))
plt.scatter(range(len(X)),GM.score_samples(X))
plt.xlim(-10,1000)
plt.xticks(range(len(X))[::20],rotation=90)
plt.plot(GM.score_samples(X),"r--")
plt.show()
print (pd.Series(GM.score_samples(X)).sort_values().head())
print

# also check for tiny clusters
print (pd.Series(GM.predict(X)).value_counts())
```

```python
g_mix = pd.concat([data2['County'], pd.Series(GM.score_samples(data2.iloc[:,1:]))], axis=1).sort_values([0], ascending=True).head(5)
```
